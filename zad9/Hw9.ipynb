```python
import tensorflow as tf
from tensorflow.keras.datasets import fashion_mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Dropout
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping
```

```python

(X_train, y_train), (X_test, y_test)= fashion_mnist.load_data()

X_train= X_train/255
X_test= X_test/255

y_train= to_categorical(y_train, 10)
y_test= to_categorical(y_test, 10)

#0.0005396387579638725
```

```python

model= Sequential()
model.add(Flatten(input_shape=(28,28)))
model.add(Dense(1024, activation="leaky_relu"))
model.add(Dropout(0.25))
model.add(Dense(512,activation="relu"))
model.add(Dropout(0.25))
model.add(Dense(256,activation="relu"))
model.add(Dropout(0.25))
model.add(Dense(128,activation="relu"))
model.add(Dropout(0.25))
model.add(Dense(64,activation="relu"))
model.add(Dropout(0.25))
model.add(Dense(10, activation='softmax'))


```

```python
model.compile(optimizer=Adam(learning_rate=0.0000556387579638725),loss='categorical_crossentropy', metrics=['accuracy'])


```

```python
model.fit(X_train, y_train, epochs=35, batch_size=38, validation_data=(X_test,y_test))
```

```python
test_loss, test_acc=model.evaluate(X_test, y_test)

print(test_loss)
print(test_acc)
```
